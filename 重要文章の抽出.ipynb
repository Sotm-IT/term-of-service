##抽出文章のカテゴリ分け

```python
# 関連　https://www.haya-programming.com/entry/2018/05/26/031355
# 入出力のパス
input_path = '/content/ToS_LINE.txt'
input_path_mon = '/content/金銭.txt'
input_path_cont = '/content/契約.txt'
input_path_info = '/content/個人情報.txt'
input_path_oth = '/content/その他.txt'
output_path_mon = '/content/result_mon.txt'
output_path_cont = '/content/result_cont.txt'
output_path_info = '/content/result_info.txt'
output_path_oth = '/content/result_oth.txt'

# search対象の取り込み
with open(input_path_mon, mode='r', encoding='utf-8') as s:
  s2 = s.readlines()
  search_target_mon =[]
  for line1 in s2:
    line2 = line1.replace('\n', '')
    search_target_mon.append(line2)

with open(input_path_cont, mode='r', encoding='utf-8') as s:
  s2 = s.readlines()
  search_target_cont =[]
  for line1 in s2:
    line2 = line1.replace('\n', '')
    search_target_cont.append(line2)

with open(input_path_info, mode='r', encoding='utf-8') as s:
  s2 = s.readlines()
  search_target_info =[]
  for line1 in s2:
    line2 = line1.replace('\n', '')
    search_target_info.append(line2)

with open(input_path_oth, mode='r', encoding='utf-8') as s:
  s2 = s.readlines()
  search_target_oth =[]
  for line1 in s2:
    line2 = line1.replace('\n', '')
    search_target_oth.append(line2)


# erase対象の設定
erase_target_list = ['」とは','日本法']

# ファイル読み込みとsearch
with open(input_path, mode='r', encoding='utf-8') as f:
    lines = f.readlines()
    # 指定した文字列を含む行を取得
    SEARCHED_TARGET = [line for line in lines if any(map(lambda x:x in line, search_target_mon)) and len(line) >= 20]
    # 指定した文字列を含む行を削除
    MON_TARGET = [line for line in SEARCHED_TARGET if all(map(lambda x:x not in line, erase_target_list))]

with open(input_path, mode='r', encoding='utf-8') as f:
    lines = f.readlines()
    #削除対象の更新
    MON_TARGET2 = []
    for line1 in MON_TARGET:
      line2 = line1.replace('\n','')
      MON_TARGET2.append(line2)
    erase_target_list = erase_target_list + MON_TARGET2
    # 指定した文字列を含む行を取得
    SEARCHED_TARGET = [line for line in lines if any(map(lambda x:x in line, search_target_info)) and len(line) >= 20]
    # 指定した文字列を含む行を削除
    INFO_TARGET = [line for line in SEARCHED_TARGET if all(map(lambda x:x not in line, erase_target_list))]

with open(input_path, mode='r', encoding='utf-8') as f:
    lines = f.readlines()
    #削除対象の更新
    INFO_TARGET2 = []
    for line1 in INFO_TARGET:
      line2 = line1.replace('\n','')
      INFO_TARGET2.append(line2)
    erase_target_list = erase_target_list + INFO_TARGET2
    # 指定した文字列を含む行を取得
    SEARCHED_TARGET = [line for line in lines if any(map(lambda x:x in line, search_target_cont)) and len(line) >= 20]
    # 指定した文字列を含む行を削除
    CONT_TARGET = [line for line in SEARCHED_TARGET if all(map(lambda x:x not in line, erase_target_list))]
 
with open(input_path, mode='r', encoding='utf-8') as f:
    lines = f.readlines()
    #削除対象の更新
    CONT_TARGET2 = []
    for line1 in CONT_TARGET:
      line2 = line1.replace('\n','')
      CONT_TARGET2.append(line2)
    erase_target_list = erase_target_list + CONT_TARGET2
    # 指定した文字列を含む行を取得
    SEARCHED_TARGET = [line for line in lines if any(map(lambda x:x in line, search_target_oth)) and len(line) >= 20]
    # 指定した文字列を含む行を削除
    OTH_TARGET = [line for line in SEARCHED_TARGET if all(map(lambda x:x not in line, erase_target_list))]

#取得した行の出力
print(MON_TARGET)
print(CONT_TARGET)
print(INFO_TARGET)
print(OTH_TARGET)

# ファイル出力
with open(output_path_mon, mode='w', encoding='utf-8') as u:
    u.writelines(MON_TARGET)

with open(output_path_cont, mode='w', encoding='utf-8') as u:
    u.writelines(CONT_TARGET)

with open(output_path_info, mode='w', encoding='utf-8') as u:
    u.writelines(INFO_TARGET)

with open(output_path_oth, mode='w', encoding='utf-8') as u:
    u.writelines(OTH_TARGET)
```
